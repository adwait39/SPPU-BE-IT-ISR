```python
def show_img(images, titles):
    # Determine the number of images
    num_images = len(images)
    
    # Create a subplot for each image
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))
    
    # If only one image is passed, axes will not be iterable, so handle that case
    if num_images == 1:
        axes = [axes]
    
    # Iterate over images and their respective titles
    for i, (image, title) in enumerate(zip(images, titles)):
        axes[i].imshow(image, cmap='gray')  # or remove cmap if images are RGB
        axes[i].set_title(title)
        axes[i].axis('off')  # Hide the axes
    
    plt.show()
```

```python
# Assuming 'red' and 'yellow' are images you already processed
image = mpimg.imread("ISRimg.jpg")
red = image[:, :, 0]  # Assuming red channel is the first
yellow = image[:, :, 1]  # Assuming yellow is a combination of channels

# Call the function to display the images
show_img(images=[red, yellow], titles=['Red Intensity', 'Yellow Intensity'])
```

```python
import matplotlib.image as mpimg
from scipy.ndimage import gaussian_filter

# Read the image
image = mpimg.imread("ISRimg.jpg")

# Convert the image to grayscale (if it's RGB)
if len(image.shape) == 3:
    grayscale_image = np.mean(image, axis=2)  # Convert to 2D by averaging RGB channels

# Apply Gaussian blur with a chosen sigma value (e.g., sigma=2)
sigma_value = 20
blurred_image = gaussian_filter(grayscale_image, sigma=sigma_value)

# Display the blurred grayscale image
plt.imshow(blurred_image, cmap='gray')
plt.title(f'Grayscale Image with Sigma Blur (σ={sigma_value})')
plt.axis('off')  # Hide the axis
plt.show()

# Print the blurred 2D matrix
print("2D Matrix Representation of Blurred Image:")
print(blurred_image)
```

```python

# Read the image
image = mpimg.imread("ISRimg.jpg")

# Convert the image to grayscale (if it's RGB)
if len(image.shape) == 3:
    grayscale_image = np.mean(image, axis=2)  # Convert to 2D by averaging RGB channels

# Apply Gaussian blur with a chosen sigma value (e.g., sigma=2)
sigma_value = 2
blurred_image = gaussian_filter(grayscale_image, sigma=sigma_value)

# Plot histograms for the original grayscale image and the blurred image
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original Grayscale Image Histogram
axes[0].hist(grayscale_image.ravel(), bins=256, range=(0.0, 1.0), color='gray')
axes[0].set_title('Histogram of Grayscale Image')
axes[0].set_xlabel('Pixel Intensity')
axes[0].set_ylabel('Frequency')

# Blurred Grayscale Image Histogram
axes[1].hist(blurred_image.ravel(), bins=256, range=(0.0, 1.0), color='gray')
axes[1].set_title(f'Histogram of Blurred Image (σ={sigma_value})')
axes[1].set_xlabel('Pixel Intensity')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
